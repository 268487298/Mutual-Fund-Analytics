```python
import pandas as pd
from pathlib import Path

root = Path.cwd()
if root.name != 'Mutual Analytics':
    root = Path('e:/Bluestock/Mutual Analytics')
raw_dir = root / 'data' / 'raw'

files = sorted(raw_dir.glob('*.csv'))
print('Found CSV files:', [p.name for p in files])

for path in files:
    print('=== {} ==='.format(path.name))
    df = pd.read_csv(path)
    print('shape:', df.shape)
    print('dtypes:')
    print(df.dtypes)
    print('head:')
    print(df.head(5).to_string(index=False))
    missing = df.isna().sum()
    if missing.any():
        print('missing counts:')
        print(missing[missing > 0].to_string())
    else:
        print('missing counts: none')
```

# validates AMFI code

```python
import pandas as pd

fund_master = pd.read_csv("../data/raw/01_fund_master.csv")
nav_history = pd.read_csv("../data/raw/02_nav_history.csv")
```

```python
print(fund_master.columns)
print(nav_history.columns)
```

```python
fund_codes = set(fund_master["amfi_code"].astype(str))
nav_codes = set(nav_history["amfi_code"].astype(str))

missing_codes = fund_codes - nav_codes

print("Fund Master Codes:", len(fund_codes))
print("NAV History Codes:", len(nav_codes))
print("Missing Codes:", len(missing_codes))
```

```python
missing_df = fund_master[fund_master["amfi_code"].astype(str).isin(missing_codes)]

missing_df.head()
```

```python
matched_codes = fund_codes.intersection(nav_codes)

match_rate = (len(matched_codes) / len(fund_codes)) * 100

print(f"Match Rate: {match_rate:.2f}%")
```
