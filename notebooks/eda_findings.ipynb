# Key EDA Findings

- NAV shows a long-term upward trajectory, with a clear 2023 rally and a 2024 correction visible in the time series.
- AUM is concentrated among a few leading fund houses, with SBI appearing as one of the largest houses across monthly data.
- Monthly SIP inflows have trended upward, reaching a December 2025 peak near ₹31,002 crore and indicating strong systematic investment momentum.
- Category net inflows are uneven across product segments and months, with the heatmap suggesting distinct seasonality and category-level strength.
- Investor transactions are dominated by a narrow set of age groups, showing a skewed age distribution in SIP participation.
- SIP amount distributions vary by age group, revealing different investment sizes and saver behavior across cohorts.
- Geographic investment is concentrated in a few states and higher-tier cities, reflecting an urban/regional concentration of mutual fund flows.
- Total folio counts have grown rapidly, roughly doubling from about 13.26 crore in January 2022 to about 26.12 crore in December 2025.
- NAV return correlations among sampled funds are strongly positive, indicating common market drivers across the selected set.
- Portfolio holdings are concentrated in the top sectors, showing a sector bias in mutual fund allocations.
