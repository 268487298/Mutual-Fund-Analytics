# performance

```python
perform = pd.read_csv('../data/raw/07_scheme_performance.csv')
perform.head()
```

```python
perform.info()
```

```python
return_cols = ["return_1yr_pct", "return_3yr_pct", "return_5yr_pct"]

for col in return_cols:
    perform[col] = pd.to_numeric(perform[col], errors="coerce")
```

```python
for col in return_cols:
    print(f"{col} - Missing values: {perform[col].isnull().sum()}")
```

```python
perform.columns
```

```python
perform[perform['sharpe_ratio'] <=0]
```

```python
expense_ratio_flag = perform[(perform['expense_ratio_pct'] > 0.1) & (perform['expense_ratio_pct'] < 2.5)]
```

```python
perform.to_csv('../data/processed/clean_scheme_performance.csv', index=False)
```

# sip inflow

```python
sip = pd.read_csv('../data/raw/04_monthly_sip_inflows.csv')
sip.head()
```

```python
sip.info()
```

```python
sip['month'] = sip['month'].astype('datetime64[ns]').dt.month
```

```python
sip.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
sip['yoy_growth_pct'] = imputer.fit_transform(sip[['yoy_growth_pct']])
```

```python
sip.isnull().sum()
```

```python
sip.duplicated().sum()
```

```python
sip.to_csv('../data/processed/clean_monthly_sip_inflows.csv', index=False)
```
